```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
#한글 폰트 설정
#font_list = font_manager.findSystemFonts(fontpaths=None,fontext="ttf")
#print(font_list)

path = 'C:\\Windows\\Fonts\\HYKANM.TTF'
font = font_manager.FontProperties(fname=path).get_name()
plt.rc('font',family=font)
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y,color="red",linestyle="--",linewidth=3,label="샘플그래프")
plt.title("매트플롯")
plt.legend(title="범례",fontsize=12,loc="lower right")
plt.show()
```

```python
plt.plot(x,y,marker="*",markersize=10,markerfacecolor="yellow",markeredgecolor="red")
plt.title("매트플롭",fontsize=10,pad=20,color="red",backgroundcolor="yellow")
```

```python
font = {
    "size" : 20,
    "color" : "white",
    "backgroundcolor" : "black",
    "weight" : "bold"
}

plt.title("매트플롭",pad=10,fontdict=font)
plt.plot(x,y,color='b')
plt.xlabel("x축",fontdict=font,labelpad=20)
plt.ylabel("y축",fontdict=font,labelpad=20)
plt.grid(True,linestyle=":",alpha=0.5,color="r")

#plt.xlim([0,5])
#plt.ylim([0,10])
#plt.axis("auto")
```

```python
#그래프 여러 개 그리기
x = [0,1,2,3,4]
y1 = [0,1,2,3,4]
y2 = [0,2,4,6,8]
y3 = [0,3,6,9,12]
y4 = [0,4,8,12,16]

plt.plot(x,y1,label="y=x",color="r",marker="o")
plt.plot(x,y2,label="y=2x",color="orange",marker="o")
plt.plot(x,y3,label="y=3x",color="green",marker="o")
plt.plot(x,y4,label="y=4x",color="b",marker="o")
plt.grid(True,linestyle=":",color="k")
plt.legend(loc="upper left",title="4개", ncol=4)
plt.title("그래프 여러 개 그리기")
plt.xlabel("x값")
plt.ylabel("y값")
```

```python
#subplot
plt.subplot(2,2,1) 
plt.plot(x,y1,"r")
plt.title("x=y")
plt.subplot(2,2,2)
plt.plot(x,y2,"orange")
plt.title("x=2y")
plt.subplot(2,2,3)
plt.plot(x,y3,"g")
plt.title("x=3y")
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("x=4y")

plt.suptitle("SubPlot")
plt.tight_layout()
```

```python
categories = ["A","B","C"]
values = [10,15,7]
bars = plt.bar(categories,values,width=0.2,color=["r","g","b"],alpha=0.5,edgecolor="k",label=['2023','2024','2025'])
plt.title("막대그래프",pad=20)
plt.xlabel("카테고리")
plt.ylabel("수치")
plt.xticks(categories,['2023','2024','2025'])
plt.legend(title="연도별")
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2,
            bar.get_height() + 0.7,
            str(bar.get_height()),
             ha="center", va="top", color="k")
```

```python
categories = ["2023","2024","2025"]
values = [10,15,7]

bars = plt.barh(categories, values, color=["r","g","b"],edgecolor="k", height = 0.4,alpha=0.5,label=["2023","2024","2025"])

for bar in bars:
    plt.text(bar.get_width() + 0.5,
             bar.get_y() + bar.get_height()/2,
             str(bar.get_width()),
             ha = "right",va="center",color="k")

plt.legend(title="연도별")
```
