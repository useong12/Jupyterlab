```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
import numpy as np
#한글 폰트 설정
#font_list = font_manager.findSystemFonts(fontpaths=None,fontext="ttf")
#print(font_list)

path = 'C:\\Windows\\Fonts\\HYKANM.TTF'
font = font_manager.FontProperties(fname=path).get_name()
plt.rc('font',family=font)
```

```python
data = [1,2,2,3,3,3,4,5,5,5,5,5,6]

plt.hist(data,bins=6,color='g',alpha=0.5)
plt.title("히스토그램")
plt.xlabel("값")
plt.ylabel("빈도수")
```

```python
data1 = [1,2,2,3,3,3,4]
data2 = [2,3,3,4,4,5,6]

plt.hist([data1,data2],bins=6,color=['r','g'],alpha=0.5,label=['data1','data2'])
plt.title("데이터 비교")
plt.xlabel("값")
plt.ylabel('빈도수')
plt.legend(title ="데이터 종류")
```

```python
#삼전도
x = [1,2,3,4,5]
y = [2,3,5,7,11]
sizes = [20,50,80,100,200]
colors = [10,20,30,40,50]

plt.scatter(x,y,s=sizes,c=colors,cmap = "viridis")
plt.colorbar(label='colorbar')
plt.show()
```

```python
n = 50
x = np.random.rand(n) # 0~1 사이 난수 값 생성
y = np.random.rand(n)

area = (30*np.random.rand(n))**2 # 0~1 사이의 난수를 생성한 뒤 30을 곱해서 제곱
color = np.random.rand(n)

plt.scatter(x,y,s=area,c=color, cmap="Spectral", alpha=0.7)
plt.show()
```

```python
# 파이
sizes = [15,30,45,10]
labels = ['사과','바나나','포도','체리']
explode = [0,0.1,0,0] # 값 강조

plt.pie(sizes,labels=labels,explode=explode,autopct="%1.1f%%", shadow=True, startangle=140)
```

```python
sizes = [10,20,30,40]
label = ['a','b','c','d']
color = ['r','g','pink','purple']

plt.pie(sizes,labels=label,colors=color,autopct="%1.1f%%",textprops={"fontsize":15, "color":"blue"},wedgeprops={"edgecolor":"black"})
```
