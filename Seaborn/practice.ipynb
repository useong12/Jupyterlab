```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

penguins = sns.load_dataset('penguins')
penguins.head(5)
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins,
            label="Penguin Species Avarage Weight", color="b")
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm",
                hue="species",
                palette="ch:r=-.2,d=.3_r",
                sizes=(1, 8), linewidth=0,
                data=penguins)
```

```python
sns.catplot(
    x='island',
    y='body_mass_g',
    data = penguins,
    hue = 'sex',
    kind = 'violin')
```

```python
flights = sns.load_dataset('flights')
flights.head(5)
```

```python
flights_mean = flights.groupby("year")["passengers"].mean() #.reset_index()를 사용해서 전에 기존 인덱스를 데이터프레임의 열로 변환
flights_mean
plt.plot(flights_mean)
```

```python
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers',observed=True)
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True,fmt=".0f",cmap="coolwarm")
```

```python
passengers_1949 = flights[flights['year']==1949]
passengers_1949
sns.barplot(x='month', y='passengers', data=passengers_1949,
            label="Passengers in 1949", color="b")
```
