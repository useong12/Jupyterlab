```python
import pandas as pd
from pydataset import data
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#print(data())
```

```python
mtcars = pd.DataFrame(data('mtcars'))
mtcars.head(5)
```

```python
graph = sns.barplot(x='cyl', y='mpg', data=mtcars, label = 'CYL-MPG')
for i in graph.patches:
    graph.text(i.get_x()+i.get_width()/2.,
               i.get_height()/2,
               f"{i.get_height():.2f}",
               ha = 'center',
               va = 'bottom')
```

```python
graph = sns.barplot(x='am', y='hp', data=mtcars, label = 'AM-HP')
for i in graph.patches:
    graph.text(i.get_x()+i.get_width()/2.,
               i.get_height()/2,
               f'{i.get_height():.2f}',
               ha = 'center',
               va = 'bottom')
```

```python
mtcars_pivot = mtcars.pivot_table(index='gear',columns='cyl',values='mpg',observed='True',fill_value=0)
plt.figure(figsize=(10, 8))
sns.heatmap(mtcars_pivot, annot=True,fmt=".2f",cmap="coolwarm",alpha=0.7)
plt.title("HEATMAP : GEAR-CYL AVG MPG")
```

```python
mtcars_three = mtcars[['mpg','hp','wt']]
matrix = mtcars_three.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(matrix, annot=True,fmt=".2f",cmap="coolwarm",alpha=0.7)
plt.title("HEATMAP : MPG-HP-WT")
```
