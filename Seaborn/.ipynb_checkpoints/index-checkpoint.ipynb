```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#산점도 그리기
tips = sns.load_dataset('tips')

sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    style="time",
    size="size",
    data = tips )
```

```python
sns.stripplot(
    x = 'day',
    y = 'total_bill',
    data = tips,
    jitter = True,
    hue = "size",
    dodge = True)
```

```python
sns.swarmplot(
    x = "day",
    y = "total_bill",
    data = tips,
    hue = "size",
    dodge = True)
```

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    style = "time",
    hue = "sex")
```
